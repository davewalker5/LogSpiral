```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Ellipse

def plot_2d_spiral(theta, x, y, aperture_width, aperture_height, figsize=10):
    """
    Plot a 2D logarithmic spiral with simulated aperture growth

    :param theta: Angle for each point
    :param x: X coordinate for each point
    :param y: Y coordinate for each point
    :param aperture_width: Simulated aperture width for each point
    :param aperture_height: Simulated aperture height for each point
    """
    _, ax = plt.subplots(figsize=(figsize, figsize))
    ax.set_aspect('equal')
    step = 40

    for i in range(0, len(theta), step):
        ellipse = Ellipse(
            (x[i], y[i]),
            width=aperture_width[i],
            height=aperture_height[i],
            angle=np.degrees(theta[i]),
            fill=False,
            linewidth=1
        )
        ax.add_patch(ellipse)

    ax.plot(x, y, alpha=0.3)
    ax.set_title("Simulated Shell Growth")
    plt.show()
```
