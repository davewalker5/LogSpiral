```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Ellipse
```

```python
def plot_2d_spiral(theta, x, y, aperture_width, aperture_height, figsize=10):
    """
    Plot a 2D logarithmic spiral with simulated aperture growth

    :param theta: Angle for each point
    :param x: X coordinate for each point
    :param y: Y coordinate for each point
    :param aperture_width: Simulated aperture width for each point
    :param aperture_height: Simulated aperture height for each point
    """
    _, ax = plt.subplots(figsize=(figsize, figsize))
    ax.set_aspect('equal')
    step = 40

    for i in range(0, len(theta), step):
        ellipse = Ellipse(
            (x[i], y[i]),
            width=aperture_width[i],
            height=aperture_height[i],
            angle=np.degrees(theta[i]),
            fill=False,
            linewidth=1
        )
        ax.add_patch(ellipse)

    ax.plot(x, y, alpha=0.3)
    ax.set_title("Simulated Shell Growth")
    plt.show()
```

```python
def plot_2d_conical_helix(theta, x, y, aperture_width, aperture_height, figsize=10):
    """
    Plot a 2D logarithmic spiral with simulated aperture growth

    :param theta: Angle for each point
    :param x: X coordinate for each point
    :param y: Y coordinate for each point
    :param aperture_width: Simulated aperture width for each point
    :param aperture_height: Simulated aperture height for each point
    """
    _, ax = plt.subplots(figsize=(figsize, figsize))
    ax.set_aspect('equal')
    step = 40

    for i in range(0, len(theta), step):
        ellipse = Ellipse(
            (x[i], y[i]),
            width=aperture_width[i],
            height=aperture_height[i],
            angle=np.degrees(theta[i]),
            fill=False,
            linewidth=1
        )
        ax.add_patch(ellipse)
    ax.plot(x, y, alpha=0.3)
    ax.set_title("Conical Helix")
    plt.show()
```

```python
def plot_2d_centreline(
    z,
    aperture_width,
    aperture_height,
    name,
    figsize=12,
    step=40
):
    """
    Plot a 2D centreline growth axis with simulated aperture expansion.

    This shows a side elevation of the shell:
    - z is the growth axis
    - aperture ellipses are drawn perpendicular to the axis
    - aperture height shows the visible taper of the shell

    :param z: Z coordinate / growth axis position for each point
    :param aperture_width: Simulated aperture width for each point
    :param aperture_height: Simulated aperture height for each point
    :param name: Name of the growth axis
    :param figsize: Width of the figure
    :param step: Interval between drawn aperture ellipses
    """

    _, ax = plt.subplots(figsize=(figsize, figsize / 3))
    ax.set_aspect("equal")

    y = np.zeros_like(z)

    for i in range(0, len(z), step):
        ellipse = Ellipse(
            (z[i], y[i]),
            width=aperture_width[i],
            height=aperture_height[i],
            angle=0,
            fill=False,
            linewidth=1
        )
        ax.add_patch(ellipse)

    ax.plot(z, y, alpha=0.3)

    upper = aperture_height / 2
    lower = -aperture_height / 2

    ax.plot(z, upper, alpha=0.5)
    ax.plot(z, lower, alpha=0.5)

    ax.set_title(f"{name} Growth")
    ax.set_xlabel("Growth axis")
    ax.set_ylabel("Aperture height")

    plt.show()
```
