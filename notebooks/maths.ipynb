```python
import numpy as np

def logarithmic_spiral(
    a=1.0,
    b=0.15,
    turns=5,
    points=2000,
    aperture_width_multiplier=0.20,
    aperture_height_multiplier=0.10,
    aperture_mode="proportional",
    abutment_scale=1.45,
    abutment_height_ratio=0.70
):
    """
    Generate coordinates, radius and simulated aperture size for a logarithmic spiral.

    :param a: Initial scale or radius
    :param b: Growth rate
    :param turns: Number of complete spiral rotations
    :param points: Number of points in the complete spiral
    :param aperture_width_multiplier: Width as proportion of r, used in proportional mode
    :param aperture_height_multiplier: Height as proportion of r, used in proportional mode
    :param aperture_mode: "proportional" or "abutting"
    :param abutment_scale: Scale factor for abutting whorl spacing
    :param abutment_height_ratio: Height/width ratio in abutting mode
    :return: theta, r, x, y, width, height
    """

    theta = np.linspace(0, turns * 2*np.pi, points)

    r = a * np.exp(b * theta)
    x = r * np.cos(theta)
    y = r * np.sin(theta)

    if aperture_mode == "proportional":
        width = aperture_width_multiplier * r
        height = aperture_height_multiplier * r

    elif aperture_mode == "abutting":
        r_inner = r * np.exp(-b * 2*np.pi)
        radial_gap = r - r_inner

        # Width is based on the spacing between neighbouring whorls.
        # This helps adjacent coils abut or overlap in a shell-like way.
        width = abutment_scale * radial_gap

        # Height is derived from width to control the aperture cross-section.
        height = abutment_height_ratio * width

    else:
        raise ValueError(
            "aperture_mode must be either 'proportional' or 'abutting'"
        )

    return theta, r, x, y, width, height
```

```python
def conical_helix(
    a=0.85,
    b=0.018,
    turns=9.5,
    points=950,
    spire_height=9.5,
    cone_taper=0.82,
    aperture_width_multiplier=0.42,
    aperture_height_multiplier=0.55
):
    """
    Generate coordinates, radius and simulated aperture size for a conical helix.

    :param a: Initial scale or radius
    :param b: Growth rate
    :param turns: Number of complete spiral rotations
    :param points: Number of points in the complete spiral
    :param spire_height: Total vertical extent of helix from apex to aperture
    :param cone_taper: Controls the vertical distribution of whorls between apex and aperture
    :param aperture_width_multiplier: Width as proportion of r, used in proportional mode
    :param aperture_height_multiplier: Height as proportion of r, used in proportional mode
    :return: theta, r, x, y, z, width, height
    """
    theta = np.linspace(0, 2*np.pi*turns, points)
    t = theta / theta[-1]

    # Radius grows slowly from apex to final whorl
    r = a * np.exp(b * theta)

    # Tower-shell axis: whorls stacked vertically
    x = r * np.cos(theta)
    y = r * np.sin(theta)

    # Cone taper < 1 : Spacing between whorls increases from apex to aperture
    # Cone taper = 1 : Whorls evenly spaced between the apex and aperture
    # Cone taper > 1 : Spacing between whorls decreases from apex to aperture
    z = -spire_height * (1 - t) ** cone_taper

    width = aperture_width_multiplier * r
    height = aperture_height_multiplier * r

    return theta, r, x, y, z, width, height
```
